```python
from pylab import *
import csv

```

```python
#agregar la parte de funciones

datos_completos = []
with open('Tarea1-Datos-aceleraci_n - Grabaci_n 3.csv') as csvfile:#exporta el archivo para leerlo
    datos_leidos = csv.reader(csvfile , delimiter =',') #asigna a una variable los datos en forma de caracteres
    for renglon in datos_leidos:
        datos_completos.append(renglon)#agrega renglon al fina de la lista
N_datos = len(datos_completos[1:])
datos_numericos = zeros((4,N_datos))#dos columnas con n datos
for i in range (N_datos):
    datos_numericos[0,i] = float(datos_completos[i+1][0])#inicia en i+1 para no intentar convertir el encabezado en flotante
    datos_numericos[1,i] = float(datos_completos[i+1][1])
    datos_numericos[2,i] = float(datos_completos[i+1][2])
    datos_numericos[3,i] = float(datos_completos[i+1][3])
plot(datos_numericos[0,:] , datos_numericos[1,:])
plot(datos_numericos[0,:] , datos_numericos[2,:])
plot(datos_numericos[0,:] , datos_numericos[3,:])

grid()
show()
```
