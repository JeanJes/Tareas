```python
def der_h4(x , f):
    f_prima = zeros_like(f) #hacemos un arreglo de puros ceros
    d_h , h = x[2] - x[0] , x[1] - x[0] #definimos los pasos o el intervalo de la derivada
    
    
    for i in range(1, len(f) - 2): #estamos dando la derivada de la funcion en el interior
        f_prima[i] = (f[i+1]-f[i-1])/d_h
         
        print('La f_prima[%i] es: %f' % (i, f_prima[i]))
    #en las dos siguientes lineas se definen los extremos  
    
       
    f_prima[0] = (-(25/12)*f[0] + 4*f[1] - 3*f[2] + (4/3)*f[3] - (1/4)*f[4])/h
    #f_prima[1] = (-(25/12)*f[1] + 4*f[2] - 3*f[3] + (4/3)*f[4] - (1/4)*f[5])/h
    #f_prima[i] = (1/12*d_h)*(-f[i+2] + 8*f[i+1] + 8*f[i-1] + f[i-2])
    f_prima[-1] = -(-(25/12)*f[-1] + 4*f[-2] - 3*f[-3] + (4/3)*f[-4] - (1/4)*f[-5])/h
    f_prima[-2] = -(-(25/12)*f[-2] + 4*f[-3] - 3*f[-4] + (4/3)*f[-5] - (1/4)*f[-6])/h
    #f_prima[-2] = (1/12*d_h)*(-f[0] + 8*f[-1] + 8*f[-3] + f[-4])
    #f_prima[-2] = -(-(25/12)*f[-2] + 4*f[-3] - 3*f[-4] + (4/3)*f[-4] - (1/4)*f[-6])/h
    #f_prima[-3] = -(-(25/12)*f[-3] + 4*f[-4] - 3*f[-5] + (4/3)*f[-6] - (1/4)*f[-7])/h

    print(f_prima[-1])
    print(f_prima[-2])
    return f_prima
```

```python
x = linspace(0, 10, 1000)
f = x**2
# derivada analitica
dy = 2*x
dy_h4 = der_h4(x , f)
```

```python
figure(dpi=100)
plot(x,dy,label="f'")
plot(x,dy_h4,label="f' numérica")
legend()
show()
```
